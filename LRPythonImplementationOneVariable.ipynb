```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(12.0, 9.0)
```

```python
data = pd.read_csv('data.csv')
data
```

```python
X = data.iloc[:, 4]
Y = data.iloc[:, 1]
plt.scatter(X, Y)
plt.title('Sales vs Sales')
plt.xlabel('Sales')
plt.ylabel('TV')
plt.show()
```

```python
#Building the model
a = 0
b = 0


L = 0.0001  # The learning Rate
epochs = 50  # The number of iterations to perform gradient descent

n = float(len(X)) # Number of elements in X

# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = a*X + b  # The current predicted value of Y
    D_a = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt a
    D_b = (-2/n) * sum(Y - Y_pred)  # Derivative wrt b
    a = a - L * D_a  # Update a
    b = b - L * D_b  # Update b

print (a, b)
```

```python
# Making predictions
Y_pred = a*X + b

plt.scatter(X, Y) 
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='green')  # regression line
plt.show()
```
